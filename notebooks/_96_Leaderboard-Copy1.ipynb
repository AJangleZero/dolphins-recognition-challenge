```python
# exporti

def get_new_df_line(zip_name):
    """Creates a new line from file name to be added to the DataFrame and then to the csv file"""
    split_list = list(zip_name.split("-"))
    date_and_time = split_list[3].split("T")
    split_list.extend(date_and_time)
    score_split = split_list[5].split("=")[1]
    split_list.append(score_split)
    split_list.append(zip_name)
    return split_list
```

```python
# exporti


def remove_unnecessary_columns(column_list, remove_column_list=[0, 2, 3, 4, 5, 8, 10]):
    """Removes unnecessary columns"""
    for r in remove_column_list:
        column_list.pop(r - remove_column_list.index(r))
    return column_list
```

```python
# exporti


def set_order(not_in_order):
    """Places the columns in the default order"""
    not_in_order[0], not_in_order[-1] = not_in_order[-1], not_in_order[0]
    not_in_order[1], not_in_order[-1] = not_in_order[-1], not_in_order[1]
    not_in_order.append(not_in_order.pop(2))
    return not_in_order
```

```python
# export


def unzip_eval_remove_save():
    """Unzip file, load model, eval model, save score, delete unzipped file and zipped at the end"""
    modelas_path = Path("models_for_evaluation")
    list_of_zip_files = [x for x in modelas_path.glob("*.zip")]

    for zip_file in list_of_zip_files:
        leaderboard_frame = pd.read_csv("leaderboard.csv")

        with zipfile.ZipFile(zip_file, "r") as zip_ref:
            zip_ref.extractall("models_for_evaluation")
        unzipped_path = [x for x in modelas_path.glob("submiss*")][0]
        print(f"Unzipped {unzipped_path}")

        new_row_list = create_new_line(zip_file.name)
        # new_row = _get_new_df_line(zip_file.name)
        if leaderboard_frame["file_name"].str.contains(new_row_list[0]).any():
                                            #contains(new_row["file_name"].any())
            print(
                f"Skip because a record of that file already exists. ({new_row_list[0]})"
            )
            continue

        print("Loading model and eval model")

        model = torch.load(unzipped_path / "model.pt")
        data_loader, data_loader_test = get_dataset("segmentation", batch_size=4)

        iou, iou_df = iou_metric(model, data_loader_test.dataset)
        print(f"IOU {iou}!")

        new_row = pd.DataFrame(
            {
                "file_name": str(new_row_list[0]),
                "year": int(new_row_list[1]),
                "month": int(new_row_list[2]),
                "day": int(new_row_list[3]),
                "time": new_row_list[4],
                "score": float(new_row_list[5]),
                "alias": str(new_row_list[6]),
                "email": str(new_row_list[7]),
            },
            index=[10000],
        )
        new_row["test_score"] = iou  # EVAL!!!!
        leaderboard_frame = leaderboard_frame.append(new_row)
        leaderboard_frame = leaderboard_frame.sort_values(
            by=["test_score"], ascending=False
        ).reset_index(
            drop=True
        )  # test_score
        leaderboard_frame.to_csv("leaderboard.csv", index=False, header=True)

        print(f"Delete {unzipped_path}")
        shutil.rmtree(unzipped_path)
```
